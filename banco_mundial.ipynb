```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



url: str = "https://api.worldbank.org/v2/country/CL/indicator/FP.CPI.TOTL.ZG?format=json&per_page=100"
response: any = requests.get(url)
data = response.json()

records = data[1]
df = pd.DataFrame([
    {"year": int(entry["date"]), "inflation": float(entry["value"])}
    for entry in records
    if entry["value"] is not None and int(entry["date"]) >= 1900
])
df = df.sort_values("year")
df
sns.set_style("whitegrid")
plt.figure(figsize=(11.25, 11.25))  # Tamaño cuadrado a 96 dpi = 1080px

sns.barplot(x="year", y="inflation", data=df, palette="magma")

plt.title("Inflación Anual en Chile\nDesde el año 1971", fontsize=18, weight='bold')
plt.xlabel("Año", fontsize=14)
plt.ylabel("Inflación (%)", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=2.0)

plt.show()
```

```python

# Tasa de incidencia de la pobreza a 8,30 dólares al día en Chile
url = "https://api.worldbank.org/v2/country/CL/indicator/SI.POV.UMIC?format=json&per_page=100"
response = requests.get(url)
data = response.json()
records = data[1]

df = pd.DataFrame([
    {"year": int(entry["date"]), "pobreza": float(entry["value"])}
    for entry in records
    if entry["value"] is not None and int(entry["date"]) >= 2000
])
df = df.sort_values("year")
df
```

```python
# Configurar estilo visual
sns.set_style("whitegrid")
plt.figure(figsize=(11.25, 11.25))  # Tamaño cuadrado para Instagram

# Crear gráfico
sns.barplot(x="year", y="pobreza", data=df, palette="rocket")

# Etiquetas y título
plt.title("Pobreza en Chile (% de población)\nUS$8.30 diarios (PPP 2017)", fontsize=18, weight='bold')
plt.xlabel("Año", fontsize=14)
plt.ylabel("Porcentaje de población", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=2.0)

# Guardar imágenes en formatos Instagram
plt.savefig("./pobreza_chile_instagram.jpg", dpi=96, bbox_inches='tight')
plt.savefig("./pobreza_chile_instagram.svg", format='svg', bbox_inches='tight')
```
